```python
#convert str into int
from sklearn.preprocessing import LabelEncoder
```

```python
species=LabelEncoder()
```

```python
df['species']=species.fit_transform(df['species'])
```

```python
df
```

```python
#seprate the features and labels
x=df[['sepal_length', 	'sepal_width', 	'petal_length', 	'petal_width']]
y=df['species']
```

```python
#split the data
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
#Random forest classifier
from sklearn.ensemble import RandomForestClassifier
```

```python
clf=RandomForestClassifier(n_estimators=100,criterion='gini',)
```

```python
clf.fit(x_train,y_train)
```

```python
clf.predict(x_test)
```

```python
clf.score(x_test,y_test)
```

```python
#most important features and least important
# Get feature importances
importances = clf.feature_importances_

# Create a series with feature importances
imp = pd.Series(importances, index=['sepal_length', 'sepal_width', 'petal_length', 'petal_width']).sort_values(ascending=False)

# Display the feature importances
print(imp)
```
